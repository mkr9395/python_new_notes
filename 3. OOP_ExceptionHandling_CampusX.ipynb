## Magic Method / Dunder Methods (\_\_x_\_)

### Dunder is short for "double underscore". 


In Python, a double underscore **(e.g. \_\_init\__)** is used as a naming convention to indicate that a variable, method, or attribute is intended to be private and should not be accessed directly from outside the class. However, it's not actually enforced as private and can still be accessed via a mangled name (e.g. \_\_classname\__ example).

These methods define the behavior of objects in Python and are used to define the behavior of basic operations like addition, comparison, indexing, etc.



For example, the __len__ magic method is called when you use the built-in len() function on an object, and the __add__ magic method is called when you use the + operator on two objects. By defining these methods in your own classes, you can customize the behavior of objects for specific purposes.

### Types of Magic / Dunder Methods

## 1. Constructors (\_\_init\__)

A constructor is a __special method that gets called automatically when an object is created from a class.__ The purpose of a constructor is to initialize the object's properties, or attributes, to their initial values.

The constructor method is always called **\_\_init\__()** and it is defined inside a class. When you create a new object from the class, the constructor is automatically called with the **self parameter,** which refers to the object being created, and any other arguments you pass to the constructor.

### 2. \_\_repr\_\_

- In Python, the __\_\_repr_\___ magic method is used to define a string representation of an object, which is used for debugging and development purposes. This representation should be unambiguous and, if possible, match the source code necessary to recreate the object. The built-in repr() function returns the string returned by the __repr__ method.

- By default, Python provides a __\_\_repr_\___ implementation that returns a string in the format <\_\_main__.ClassName object at 0x7f8b6c7f9d90>. However, it's common to override this method in custom classes to return a more meaningful representation of the object's state.

### 3. \_\_str\__

- In Python, the **\_\_str\__** magic method is used to define a human-readable string representation of an object. The built-in str() function returns the string returned by the **\_\_str\__** method. If **\_\_str\__** is not defined in a class, Python will use the result of **\_\_repr\__** instead.



- **\_\_str\__** should provide a concise and informative representation of the object, which is suitable for display to an end user. It should be used for the "informal" string representation of an object, as opposed to **\_\_repr\__**, which provides a more complete and unambiguous representation of an object, suitable for debugging and development.

### List of some commonly used and useful dunder methods in Python:

1. **`__init__(self, ...)`**: Initialize a new object.


2. **`__str__(self)`**: Return a human-readable string representation of the object.


3. **`__repr__(self)`**: Return an unambiguous string representation for debugging.


4. **`__len__(self)`**: Return the length of the object.


5. **`__getitem__(self, key)`**: Allow indexing, like `obj[key]`.


6. **`__setitem__(self, key, value)`**: Allow assignment to indexed values, like `obj[key] = value`.


7. **`__delitem__(self, key)`**: Allow deletion of indexed values, like `del obj[key]`.


8. **`__iter__(self)`**: Return an iterator object for iteration.


9. **`__next__(self)`**: Return the next value in the iteration.


10. **`__contains__(self, item)`**: Check if an item is present in the object, like `item in obj`.


11. **`__eq__(self, other)`**: Define the behavior of the equality operator (`==`).


12. **`__ne__(self, other)`**: Define the behavior of the inequality operator (`!=`).


13. **`__lt__(self, other)`**: Define the behavior of the less-than operator (`<`).


14. **`__gt__(self, other)`**: Define the behavior of the greater-than operator (`>`).


15. **`__le__(self, other)`**: Define the behavior of the less-than-or-equal-to operator (`<=`).


16. **`__ge__(self, other)`**: Define the behavior of the greater-than-or-equal-to operator (`>=`).


17. **`__add__(self, other)`**: Define the behavior of the addition operator (`+`).


18. **`__sub__(self, other)`**: Define the behavior of the subtraction operator (`-`).


19. **`__mul__(self, other)`**: Define the behavior of the multiplication operator (`*`).


20. **`__call__(self, ...)`**: Make an instance callable, like a function.


````python
class Fraction:
    
    # parameterized constructor
    def __init__(self,x,y):
        self.num = x
        self.den = y
    
    
    def __str__(self):
        return f'{self.num}/{self.den}'
    
    
    def __add__(self,other):
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den*other.den

        return '{}/{}'.format(new_num,new_den)
    
    
    def __sub__(self,other):
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den*other.den

        return '{}/{}'.format(new_num,new_den)
    
    
    def __mul__(self,other):
        new_num = self.num*other.num
        new_den = self.den*other.den
        
        return '{}/{}'.format(new_num,new_den)
    
    
    
    def __truediv__(self,other):
        new_num = self.num*other.den
        new_den = self.den*other.num

        return '{}/{}'.format(new_num,new_den)
    
    
    def convert_to_decimal(self):
        return self.num/self.den
````

##### 6. Two lines will intersect each other or not:

Find slope of both lines.

- If the slopes are different, the lines intersect at a single point.
- If the slopes are equal and the y-intercepts are different, the lines are parallel and do not intersect.
- If the slopes are equal and the y-intercepts are the same, the lines are coincident and intersect at every point.

##### writing a class to equation of line provied 2 sets of points. eg : points (2, 3) and (5, 7) 

````python
def do_lines_intersect(line1, line2):
    # Unpack line coordinates
    x1, y1 = line1[0]
    x2, y2 = line1[1]
    x3, y3 = line2[0]
    x4, y4 = line2[1]

    # Calculate slopes and y-intercepts
    slope1 = (y2 - y1) / (x2 - x1) if x2 - x1 != 0 else float('inf')
    slope2 = (y4 - y3) / (x4 - x3) if x4 - x3 != 0 else float('inf')
    y_intercept1 = y1 - slope1 * x1 if slope1 != float('inf') else x1
    y_intercept2 = y3 - slope2 * x3 if slope2 != float('inf') else x3

    # Check if lines intersect
    if slope1 != slope2:
        return True
    elif slope1 == slope2 and y_intercept1 == y_intercept2:
        return True
    else:
        return False
````

## Difference between Pass by reference and Pass by value

- 1. __pass by value__ means that a copy of the variable's value is passed to a function, while __pass by reference__ means that a reference or pointer to the variable's location in memory is passed to the function.



- 2. In __pass by value__, the function cannot modify the original variable outside the function. In __pass by reference__, the function can modify the original variable outside the function.


In Python, all variables are actually references to objects in memory, so technically speaking, Python uses pass by object reference. The behavior of the pass depends on the type of the object. Immutable objects like numbers and strings are effectively passed by value, while mutable objects like lists and dictionaries are effectively passed by reference.

#### Pass by Value Example (using an immutable type)

````python
def modify_number(num):
    num = 20  # This only changes the local copy of `num`, not the original
    print(id(num))
    print("Inside modify_number:", num, id(num))  # Output: 20

original_number = 10
print("Before modify_number:", original_number, id(original_number))  # Output: 10
modify_number(original_number)
print("After modify_number:", original_number, id(original_number))  # Output: 10
````

#### Pass by Reference Example (using a mutable type)

````python
def modify_list(my_list):
    my_list.append(4)  # Modifies the original list
    print(id(my_list))
    print("Inside modify_list:", my_list, id(my_list))  # Output: [1, 2, 3, 4]

original_list = [1, 2, 3]
print("Before modify_list:", original_list,id(original_list))  # Output: [1, 2, 3]
modify_list(original_list)
print("After modify_list:", original_list, id(original_list))  # Output: [1, 2, 3, 4]
````

### Explanation:

- **Pass by Value**: In the `modify_number` function, `num` is an integer (immutable type). Changing `num` inside the function does not affect `original_number` outside because integers are passed by value.


- **Pass by Reference**: In the `modify_list` function, `my_list` is a list (mutable type). Modifying `my_list` inside the function directly affects `original_list` because lists are passed by reference.

### Difference between Static and Instance variable

In object-oriented programming, variables can be classified as either static or instance variables, also known as class variables and member variables respectively.


The main differences between these two types of variables are as follows:

- Scope: Static variables are shared across all instances of a class and can be accessed without creating an object of that class. They have class-level scope and are created when the class is loaded into memory. Instance variables, on the other hand, are specific to each instance of the class and can only be accessed through an object of that class.


- Memory allocation:
Static variables are allocated memory only once, at the time the class is loaded into memory. They exist throughout the entire lifetime of the program, until the program terminates. Instance variables are allocated memory for each object that is created from the class. The memory for instance variables is allocated when an object is created and is released when the object is destroyed.


- Usage:
Static variables are typically used to store data that is common to all instances of a class, such as constants, shared data, or configuration parameters. Instance variables, on the other hand, are used to store data that is specific to each instance of a class, such as the state or properties of an object.

````python

class Employee:
    company_name = "TechCorp"  # Static variable (shared by all instances)

    def __init__(self, name, salary):
        self.name = name          # Instance variable (unique to each instance)
        self.salary = salary      # Instance variable (unique to each instance)

# Creating instances
emp1 = Employee("Alice", 50000)
emp2 = Employee("Bob", 60000)

# Accessing static variable
print(emp1.company_name)  # Output: TechCorp
print(emp2.company_name)  # Output: TechCorp

# Modifying instance variables
emp1.salary = 55000
print(emp1.salary)        # Output: 55000 (changed only for emp1)
print(emp2.salary)        # Output: 60000 (unchanged for emp2)

# Modifying static variable
Employee.company_name = "GlobalTech"
print(emp1.company_name)  # Output: GlobalTech (reflected across all instances)
print(emp2.company_name)  # Output: GlobalTech
````

### Key Differences:
- **Static variable**: Shared among all instances (e.g., `company_name`). Changing it affects all instances.
- **Instance variable**: Unique to each instance (e.g., `name` and `salary`). Changing it affects only that specific instance.

## Static Method :

The `@staticmethod` decorator in Python is used to define a method that belongs to a class but doesn’t access or modify any instance or class-specific data. It essentially makes the method independent of any instance or class-level data, acting more like a regular function that's logically related to the class.

##### Benefits of Using `@staticmethod`:
1. **Logical Grouping**: It groups utility functions with the class that logically relate to it, even though they don’t need access to instance or class variables. This improves organization and readability.


2. **No Need for `self` or `cls`**: A `staticmethod` doesn’t take `self` or `cls` as a parameter, making it clear that it doesn’t modify instance or class state. This distinction can reduce bugs caused by unintentional state modifications.


3. **Callable Without an Instance**: Static methods can be called using the class name directly without creating an instance, which can be convenient for utility or helper functions.

##### to access the get_counter() we need to use Atm.get_counter() [classname.get_counter()] as it doesnot uses self.

### Static method is a type of utility function and we add a static method decorator before the method.

### NOTE : We don't need to create objects for them, they can be directly accessed through the classname.methodname

````python
class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b
    
    @staticmethod
    def multiply(a, b):
        return a * b

# Calling static methods without creating an instance
print(MathUtils.add(5, 3))         # Output: 8
print(MathUtils.multiply(4, 6))    # Output: 24
````

In this example, `add` and `multiply` don’t depend on any class or instance variables, so they’re defined as `@staticmethod`s. They’re grouped within `MathUtils` to organize them logically, making the code more readable.

## **` What is the difference between a class method, a static method and an instance method?`**



````python

class MyClass:

    def method(self):
        return 'instance method called', self

    @classmethod
    def classmethod(cls):
        return 'class method called', cls

    @staticmethod
    def staticmethod():
        return 'static method called'

    
````

**`Instance Methods`**

The first method on `MyClass`, called `method`, is a regular instance method. That’s the basic, no-frills method type you’ll use most of the time. You can see the method takes one parameter, self, which points to an instance of MyClass when the method is called. But of course, instance methods can accept more than just one parameter.

Through the self parameter, instance methods can freely access attributes and other methods on the same object. This gives them a lot of power when it comes to modifying an object’s state.

Not only can they modify object state, instance methods can also access the class itself through the `self.__class__` attribute. This means instance methods can also modify class state.  This makes instance methods powerful in terms of access restrictions—they can freely modify state on the object instance and on the class itself.

**`Class Methods`**

Let’s compare that to the second method, `MyClass.classmethod`. I marked this method with a [`@classmethod`](https://docs.python.org/3/library/functions.html#classmethod) decorator to flag it as a class method. Instead of accepting a self parameter, class methods take a `cls` parameter that points to the class—and *not* the object instance—when the method is called.

Since the class method only has access to this `cls` argument, it can’t modify object instance state. That would require access to self. However, class methods can still modify class state that applies across all instances of the class.

#### `Static Methods`

The third method, `MyClass.staticmethod` was marked with a [`@staticmethod`](https://docs.python.org/3/library/functions.html#staticmethod) decorator to flag it as a static method.

This type of method doesn’t take a `self` or a `cls` parameter, although, of course, it can be made to accept an arbitrary number of other parameters.

As a result, a static method cannot modify object state or class state.
Static methods are restricted in what data they can access—they’re primarily a way to namespace your methods.

**Let’s See Them in Action!**

Let’s take a look at how these methods behave in action when we call them. We’ll start by creating an instance of the class and then calling the three different methods on it.

`MyClass` was set up in such a way that each method’s implementation returns a tuple containing information we can use to trace what’s going on and which parts of the class or object that method can access.


````python
# Class

class MyClass:

    def method(self):
        return 'instance method called', self

    @classmethod
    def classmethod(cls):
        return 'class method called', cls

    @staticmethod
    def staticmethod():
        return 'static method called'
````

Here’s what happens when we call an instance method:

````python
obj = MyClass()
obj.method()
````

This confirms that, in this case, the instance method called method has access to the object instance (printed as `<MyClass instance>`) via the self argument.

When the method is called, Python replaces the `self` argument with the instance object, `obj`.

We could ignore the syntactic sugar provided by the `obj.method()` **dot-call syntax** and pass the instance object manually to get the same result:
```
MyClass.method(obj)
```


````python
MyClass.method(obj)
````

Let’s try out the **class method** next:


````python
obj.classmethod()
````

Calling `classmethod()` showed us that it doesn’t have access to the `<MyClass instance>` object, but only to the `<class MyClass>` object, representing the class itself (**everything in Python is an object, even classes themselves**).

Notice how Python automatically passes the class as the first argument to the function when we call `MyClass.classmethod()`. Calling a method in Python through the **dot syntax** triggers this behavior. The `self` parameter on instance methods works the same way.

Please note that naming these parameters `self` and `cls` is just a convention. You could just as easily name them `the_object` and `the_class` and get the same result. All that matters is that they’re positioned first in the parameter list for that particular method.

Time to call the **static method** now:


````python
obj.staticmethod()
````

Did you see how we called `staticmethod()` on the object and were able to do so successfully? Some developers are surprised when they learn that it’s possible to call a static method on an object instance.

Behind the scenes, Python simply enforces the access restrictions by
not passing in the `self` or the `cls` argument when a static method gets
called using the dot syntax

This confirms that static methods can neither access the object instance state nor the class state. They work like regular functions but belong to the class’ (and every instance’s) namespace.

Now, let’s take a look at what happens when we attempt to call these
methods on the class itself, **without creating an object instance** beforehand:

````python
# Class Method
print(MyClass.classmethod())
# Static method
print(MyClass.staticmethod())
#Instance Method
print(MyClass.method())
````

We were able to call `classmethod()` and `staticmethod(`) just fine, but attempting to call the instance method `method()` failed with a `TypeError`.

This is to be expected. This time we didn’t create an object instance and tried calling an instance function directly on the class blueprint itself. This means there is no way for Python to populate the `self` argument and therefore the call fails with a `TypeError exception`.

This should make the distinction between these three method types a
little more clear

**Key Takeaways**
* Instance methods need a class instance and can access the instance through `self`.
* Class methods don’t need a class instance. They can’t access the
instance (`self`) but they have access to the class itself via `cls`.
* Static methods don’t have access to `cls or self`. They work like
regular functions but belong to the class’ namespace.
* Static and class methods communicate and (to a certain degree)
enforce developer intent about class design. This can have definite maintenance benefits.

In this example, the ParentClass has 4 attributes: attr1, attr2, attr3, and attr4. The ChildClass only wants to inherit attr1, attr2, and attr3. To achieve this, we define the __init__ method in the ChildClass and call the parent class's __init__ method using super().__init__(attr1, attr2, attr3, None). __We pass None as the value for the attr4 argument to indicate that the ChildClass does not need to inherit this attribute.__

Note that in the ChildClass, the attr4 attribute is not explicitly defined. If you try to access attr4 on an instance of ChildClass, you will get an AttributeError.

Method Resolution Order is referred to as MRO.
A class inherits from many classes under multiple inheritance.
If we attempt to access a method by building an object from the child class, the methods of the child class are first searched for the method.
If the method is not found in the child class, the inheritance classes are searched from left to right.

The `show` method is present in both the `Father` and `Mother` classes in the example presented below.

In MRO, methods and variables are searched from left to right because while conducting inheritance, `Father` class is written first and `Mother` class is written afterwards. So firstly `Father` class will be searched for `show` method if found then will get executed if not, `Mother` class will be searched.

### Types of Polymorphism : 

- __1. Method Overriding__ : method overriding is a form of polymorphism where a subclass provides a specific implementation for a method that is already defined in its superclass. This allows objects of different classes to be treated as objects of a common base. __Child and parent class have same methods and child method is used__

````python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")
        super().speak()

# Using method overriding
dog = Dog()
cat = Cat()

dog.speak()  # Output: Dog barks
cat.speak()  # Output: Cat meows and Animal speaks
````

- __2. Method Overloading__ : method overloading is achieved through default values for function parameters or using variable-length argument lists __(\*args or \**kwargs).__

````python
class MathOperations:
    def add(self, a, b=0, c=0):
        return a + b + c

obj = MathOperations()
print(obj.add(1))       # Output: 1
print(obj.add(1, 2))    # Output: 3
print(obj.add(1, 2, 3)) # Output: 6
````

````python
class MathOperations:
    def add(self, *args):
        return sum(args)

obj = MathOperations()
print(obj.add(1))       # Output: 1
print(obj.add(1, 2))    # Output: 3
print(obj.add(1, 2, 3)) # Output: 6
````

- The goal of abstraction is to simplify complex systems by hiding the underlying details and presenting a clean, easy-to-use interface. This helps to make code more manageable and reduces the risk of errors.


- abstraction can be achieved through the use of __abstract base classes (abc)__, which define the interface that must be implemented by any concrete subclasses.


__NOTE: Abstract class must have 1 @abstractmethod__

````python
from abc import ABC, abstractmethod

class AbstractClass(ABC):
    @abstractmethod
    def do_something(self):
        pass

class ConcreteImplementation(AbstractClass):
    def do_something(self):
        print("Doing something concrete.")
````

- In this example, the abstract base class AbstractClass defines an abstract method do_something that must be implemented by any concrete subclasses, such as ConcreteImplementation. This allows for a level of abstraction that makes it possible to write more flexible and reusable code.



- Abstraction is a key tool for managing complexity in object-oriented programming and helps to create a clear separation between the essential features of an object and the underlying details of its implementation.

### Assertion in Exception handling

- An assertion is a statement that __checks whether a given condition is true or false.__ 


- If the assertion condition is true, the program continues to execute normally. However, if the condition is false, an `AssertionError` exception is raised, interrupting the program's execution.


- Assertions are mainly used for debugging and testing purposes to catch logical errors and ensure that certain conditions are met during program execution. They are not intended for handling runtime errors or exceptions that might occur during normal program operation.

### Basic Syntax:
```python
assert condition, message
```
- `condition`: The expression that should evaluate to `True` or `False`.
- `message` (optional): An optional message to be displayed if the assertion fails.

### Examples:

#### Example 1: Basic Assertion

````python
x = 5
assert x > 0, "x should be positive"
print("This will be executed if the assertion is true.")
````

In this example, the assertion checks if `x` is greater than 0. Since `x` is 5, the assertion is true, and the program continues to execute normally.

#### Example 2: Assertion Failure

````python
y = -3
assert y > 0, "y should be positive"
print("This will not be executed if the assertion fails.")
````

In this example, the assertion checks if `y` is greater than 0. However, since `y` is -3, the assertion is false, and an `AssertionError` will be raised with the specified message.

#### Example 3: Using Assertions in Functions

````python
def divide(a, b):
    try:
        assert b != 0, "Cannot divide by zero"
        result = a / b
        return result
    except AssertionError as e:
        print(f"AssertionError: {e}")
        return None
    except Exception as ex:
        print(f"An unexpected error occurred: {ex}")
        return None

# Example usage with valid division
result1 = divide(10, 2)
print("Result 1:", result1)

# Example usage with division by zero (assertion failure)
result2 = divide(5, 0)
print("Result 2:", result2)
````

### Best Practices:


1. **Use Assertions for Debugging and Testing**: Assertions are primarily meant for debugging and testing. They help you catch logical errors early in the development process.


2. **Don't Use Assertions for Input Validation**: Assertions can be disabled globally, and their purpose is not to handle invalid input from users. Use proper error handling mechanisms for input validation.


3. **Keep Assertions Simple**: Assertions should be simple checks that are expected to be true during normal program execution. Complex conditions or side effects in assertions can make code hard to understand.


4. **Avoid Using Assertions for Business Logic**: Assertions are not meant to check business logic or enforce program requirements. For those cases, use proper error handling and validation techniques.


5. **Enable/Disable Assertions**: By default, assertions are enabled. You can disable them globally using the `-O` (optimize) command-line switch or the `PYTHONOPTIMIZE` environment variable. Disabling assertions in production can be useful for performance reasons but should be done cautiously.

```bash
python -O my_script.py
```

In summary, assertions are a powerful tool for catching logical errors during development and testing. However, they should be used judiciously, and their purpose should align with debugging and testing rather than handling runtime errors.
